<b>Affinity Propagation is another clustering algorithm that falls under the category of unsupervised learning.</b>

*   <b>Unlike K-means, Affinity Propagation does not
    require the number of clusters (K) to be specified in advance.</b>

*   <b>Instead, it automatically determines the number of
    clusters based on the data.</b>

*   <b>The algorithm is particularly useful when the
    number of clusters is not known beforehand or when the data has complex structures.</b>

<b>Here's an overview of how Affinity Propagation works:</b>

<b>Similarity Matrix:</b>

*   Compute a similarity matrix that quantifies how well
    each data point agrees with every other data point.

<b>Affinity Propagation:</b>

*   Initialize a "responsibility" matrix and an
    "availability" matrix.

*   Update the matrices iteratively based on messages
    passed between data points.

*   Messages indicate how well-suited a point is to be an
    exemplar for another.

<b>Exemplars:</b>

*   Identify data points as exemplars, which are
    representative members of clusters.

<b>Result:</b>

*   Each data point is assigned to the cluster associated
    with its chosen exemplar.

*   One of the key features of Affinity Propagation is
    that it can identify exemplars (cluster representatives) that are not necessarily data points themselves.

*   This property makes it suitable for tasks where
    clusters may not have a clear center.

## Import Necessary Libraries

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AffinityPropagation
```

## Make synthetic dataset

```python
x,clusters=make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=0)
#x, clusters
```

## Model Building

```python
af=AffinityPropagation(preference=-50)
clustering=af.fit(x)
```

## Visualization using scatter plot

```python
plt.scatter(x[:, 0], x[:, 1], c=clustering.labels_,cmap='rainbow', alpha=0.70, edgecolors='b')
```
