## Import Libraries

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import SpectralClustering
## make_blob is used to create a synthetic 2-dimensional data
```

## Create a blob of 200 data points

```python
dataset,clusters= make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6,random_state=50)
```

## Create ScatterPlot of original data

```python
plt.scatter(dataset[:,0],dataset[:,1])
plt.show()
```

# Perform clustering using Spectral Clustering

```python
sc=SpectralClustering(n_clusters=4).fit(dataset)
labels=sc.labels_
plt.scatter(dataset[:,0], dataset[:,1], c=labels)
plt.show()
```

## Check clustering by changing the number of clusters

```python
f=plt.figure()
f.add_subplot(2,2,1)
for i in range(2,6):
  sc=SpectralClustering(n_clusters=i).fit(dataset)
  f.add_subplot(2,2, i-1)
  plt.scatter(dataset[:,0], dataset[:,1], c=sc.labels_, label='n_cluster-' + str(i))
  plt.legend()

plt.show()
```
