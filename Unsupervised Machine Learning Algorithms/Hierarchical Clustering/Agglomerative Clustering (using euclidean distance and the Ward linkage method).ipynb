<b> Agglomerative clustering is a hierarchical clustering algorithm widely used in data analysis and machine learning. Unlike partitioning algorithms like k-means, agglomerative clustering builds a hierarchy of clusters by successively merging or agglomerating individual data points or small clusters. Here's a summary of the key aspects of agglomerative clustering:</b>

---

<b>Bottom-Up Approach:</b>

  Agglomerative clustering starts with each data point as a singleton cluster and iteratively merges the closest (most similar) pairs of clusters until only one cluster, the root, remains.

<b>Distance (Similarity) Measure:</b>

* The choice of a distance or similarity measure between clusters or data
  points is essential. Common measures include Euclidean distance, Manhattan distance, and cosine similarity.

<b>Linkage Criteria</b>:

* The linkage criterion determines how the distance between clusters is
  calculated during the merging process. Common linkage criteria include:

<b>Single Linkage:</b> Minimum distance between points in two clusters.

<b>Complete Linkage:</b> Maximum distance between points in two clusters.

<b>Average Linkage:</b> Average distance between points in two clusters.

<b>Ward's Method:</b> Minimizes the variance within clusters.

<b>Dendrogram:</b>: Agglomerative clustering produces a dendrogram, a tree-like structure that represents the hierarchy of clusters. The height of the dendrogram at a merge step represents the distance (or dissimilarity) at which the clusters were merged.

<b>Number of Clusters:</b>

*   The choice of the number of clusters is flexible and can be determined by cutting the dendrogram at a certain height or depth. This allows for hierarchical clustering at different levels.

<b>Advantages</b>:

* Hierarchical structure provides insights into the relationships between
  clusters at different scales.

* No need to specify the number of clusters beforehand.

<b>Disadvantages:</b>

* Computationally more intensive than some other clustering methods, especially
  for large datasets.
  
* Sensitivity to the choice of distance metric and linkage criteria.

<b>Applications:</b>:

* Biology (e.g., taxonomy and gene expression analysis).
* Document clustering.
* Image segmentation.
* Social network analysis.

<b>Variants</b>:

* Various agglomerative clustering variants exist, adapting the algorithm for
  specific scenarios or improving its efficiency.

## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
## make_blob is used to create a synthetic 2-dimensional data
```

## Create a blob of 200 data points

```python
dataset= make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6,random_state=50)
```

```python
dataset
```

```python
points=dataset[0]
points
```

## Create dendrogram

```python
dendrogram=sch.dendrogram(sch.linkage(points,method='ward'))
```

```python
plt.scatter(dataset[0][:,0],dataset[0][:,1])
```

# Perform the actual clustering

```python
hc=AgglomerativeClustering(n_clusters=4, affinity='euclidean',linkage='ward')
```

```python
y_hc=hc.fit_predict(points)
```

```python
plt.scatter(points[y_hc==0,0], points[y_hc==0,1], s=100,c ='cyan')
plt.scatter(points[y_hc==1,0], points[y_hc==1,1], s=100,c ='yellow')
plt.scatter(points[y_hc==2,0], points[y_hc==2,1], s=100,c ='red')
plt.scatter(points[y_hc==3,0], points[y_hc==3,1], s=100,c ='green')
```
